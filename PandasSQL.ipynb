```python
# Get necessary modules, bring data in
import pandas as pd
# PandaSQL runs on Sqlite
from pandasql import sqldf as sd
cars = pd.read_csv("Documents/cars.csv")
cars.head()
```

```python
# Initial query using a double filter
# Don't love filtering in Pandas? Use this
q = "SELECT * FROM cars WHERE mpg >= 20 AND hp >= 100"
# cars[(cars['mpg'] >= 20) & cars['hp'] >= 100]
sd(q)
```

```python
# Rename bad column name - SQL is more strict than Python on this
cars = cars.rename(columns = {'Unnamed: 0': 'model'})
```

```python
# Wildcard query
q = "SELECT * FROM cars WHERE model LIKE 'm%'"
sd(q)
```

```python
# Aggregation query
q = "SELECT cyl, MAX(mpg) AS BestMPG FROM cars GROUP BY cyl"
sd(q)
```

```python
# Bring in sales data and join the two
sales = pd.read_csv("Documents/cars_sales.csv")
sales.head()
```

```python
q = "SELECT cars.model, mpg, cyl, sales FROM cars LEFT JOIN sales ON cars.model = sales.model"
sd(q)
```
