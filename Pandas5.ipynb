```python
# Import packages and cars dataset
import pandas as pd
cars = pd.read_csv("cars_dirty.csv")
```

```python
# Let's take a look at the data
cars
```

```python
# Data is clearly missing from the cylinder column
# Options available:
# Remove the entire column
# Fill the NaNs with the preceding number
# Fill the NaNs with the average
```

```python
# Remove the entire column
cars2 = cars.drop(['cyl'], axis = 1)
cars2
```

```python
# Fill the NaNs with the preceding number
#ffill = Forward fill
cars3 = cars.ffill(axis = 0)
cars3
```

```python
# Fill the NaNs with the average
cars4 = cars.fillna(cars['cyl'].mean())
# Mean is 6, so this code becomes cars4 = cars.fillna(6)
cars4
```

```python
cars.tail(5)
```

```python
# What about duplicates?
# cars.duplicated()
cars[cars.duplicated()]
```

```python
# These could be edited or removed, depending on inclination
cars5 = cars.drop_duplicates()
cars5
```

```python
# String to lower
cars6 = cars
cars6['model'] = cars6['model'].str.lower()
cars6
```

```python
# Mapping a new column
speed = {
    4: 'Regular',
    6: 'Fast',
    8: 'Fast'
}
cars['speed'] = cars['cyl'].map(speed)
cars.head()
```

```python
# Copying/subsetting a DataFrame
# Just filter and create a new DataFrame based on the filter!
cars_fast = cars[cars['speed'] == 'Fast']
cars_fast
```

```python
# Creating pivot tables in Pandas
cars_mpg_avg = pd.pivot_table(cars, index='cyl', values='mpg', aggfunc='mean')
# Aggfunc = "aggregation function"
cars_mpg_avg
```

```python
# Joins
# Re-read in cars dataframe, to overwrite any changes
# Important to know in case you get stuck!
cars = pd.read_csv("cars_dirty.csv")
cars_sales = pd.read_csv("cars_sales.csv")
cars = cars.set_index('model')
cars_joined = cars.join(cars_sales.set_index('model'), on = 'model', how = 'left')
cars_joined
```
