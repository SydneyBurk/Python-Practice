```python
import pandas as pd
iris=pd.read_csv('iris_dirty.csv')
```

```python
iris
```

```python
# The first five values in petal_width are blank. In a new dataframe, create a syntax that will remove the petal_width column.
iris2=iris.drop('petal.width',axis=1)
iris2
```

```python
# In a new dataframe, create a syntax that replaces the blank values with the average petal width for Setosa varieties.
setosa = iris[iris['variety']=='Setosa']
setosa['petal.width'].mean()
iris3=iris.fillna(setosa['petal.width'].mean())
iris3
```

```python
# Create a syntax that will identify where duplicate values are within the dataframe.
iris[iris.duplicated()]
```

```python
# Create a syntax that will drop duplicates from the data output.
iris4=iris.drop_duplicates()
iris4
```

```python
iris.variety.unique()
```

```python
# Create a new column called IsSetosa that outputs "Setosa" if the variety is Setosa and "Not Setosa" otherwise
IsSetosa={
    'Setosa':'Setosa',
    'Versicolor':'Not Setosa',
    'Virginica':'Not Setosa'
}
iris['IsSetosa'] = iris['variety'].map(IsSetosa)
iris
```

```python
# Create a pivot table that lists the maximum sepal width for each variety.
iris_max_sepal_width=pd.pivot_table(iris, index='variety', values='sepal.width', aggfunc='max')
iris_max_sepal_width
```
